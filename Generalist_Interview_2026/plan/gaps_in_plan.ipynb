## Add before interview

These are the gaps most likely to matter:

### 1. Runtime vs persistent config changes

Be able to explain:

* temporary change now vs survives reboot
* `sysctl -w` vs config in `/etc/sysctl.conf` or `/etc/sysctl.d/`
* changing a running service/session vs changing its config file

What to say:

> Runtime changes affect the current system state now. Persistent changes are written to config so they remain after reboot or service restart.

### 2. Memory pressure and swap

Know:

* what swap is
* why swapping hurts performance
* symptoms: slowness, high I/O wait, paging, possible OOM issues

What to say:

> If memory is tight, the system may swap pages to disk. That increases latency a lot because disk is far slower than RAM.

### 3. Non-disruptive process introspection

Add comfort with:

* `ps aux`
* `top` / `htop`
* `lsof -p <pid>`
* `/proc/<pid>/`
* `strace -p <pid>` at a high level

What to say:

> I start with non-invasive inspection to understand CPU, memory, open files, sockets, and what the process is waiting on before making changes.

### 4. HTTP methods and important headers

Know clearly:

* GET, POST, PUT, DELETE
* safe/idempotent at a basic level
* `Host`, `Content-Type`, `Accept`, `Authorization`, `Cache-Control`, `Set-Cookie`

What to say:

> I first validate method, path, headers, auth, payload, and expected status code before assuming the backend is broken.

### 5. API troubleshooting ladder

Be ready to isolate failures in this order:

* DNS
* TCP connect
* TLS
* HTTP status
* auth
* payload / app logic
* rate limiting / dependency timeout

What to say:

> I narrow external API issues layer by layer so I can tell whether it is a name resolution, connectivity, TLS, authorization, or application problem.

## Top short answers to memorize

* **502**: current component is up, but upstream failed
* **Connection refused**: target reached, but nothing is listening there
* **Timeout**: traffic did not complete in time, maybe network path, firewall, or overloaded service
* **401 vs 403**: unauthenticated vs authenticated but not allowed
* **Self-signed cert**: not trusted by default because it is not signed by a trusted CA
* **Cookie**: server-managed client state
* **Cache-Control**: caching instructions, not a guarantee of caching
* **Persistent config**: survives reboot; runtime config does not
