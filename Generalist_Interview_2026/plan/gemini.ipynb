It is completely understandable to feel overwhelmed by "Generalist" infrastructure roles. When you specialize in something like Kubernetes, you get used to looking at the world through a specific set of abstractions.

The good news? **Kubernetes is just Linux, Networking, and APIs in a trench coat.** You already know this stuff; we just need to shift your lens from `kubectl` to the underlying "bare metal" and protocol logic.

Since the interview is early next week, here is your high-ROI, 4-day "No-Fluff" plan.

-----

## The "Outside-In" Strategy

In a Google TSE interview, you aren't just answering trivia; you are solving a puzzle. Always troubleshoot from the **Outside (Client/DNS)** to the **Inside (Linux Kernel/Process)**.

### Day 1: Linux Infrastructure & Systems Admin

**Focus:** How does a server breathe, and how do we check its pulse?

  * **Runtime vs. Persistent:** Understand `/etc/sysctl.conf` and the `sysctl` command. If you change a kernel parameter (like `net.core.somaxconn`) at runtime, it disappears on reboot unless written to a config file.
  * **The "Golden Four" Metrics:** CPU, Memory, Disk I/O, and Network.
      * *Tooling:* Know how to read `top` (load averages), `vmstat` (context switches/memory), and `iostat`.
      * *Memory:* Understand **paging vs. swapping**. Swapping is a performance killer.
  * **Process Introspection:** If a process is "stuck," how do you look inside without killing it?
      * `lsof -p <pid>`: What files/sockets is it holding?
      * `strace -p <pid>`: What system calls is it making right now? (Essential for TSE roles).

### Day 2: Networking & Transport Layer

**Focus:** How does the packet get from Point A to Point B?

  * **DNS Lifecycle:** This is a classic Google question. Be able to explain: Client -\> Recursive Resolver -\> Root -\> TLD -\> Authoritative. Mention **TTL** and **Caching** at every layer.
  * **TCP vs. UDP:** Apigee is an API gateway; it lives and dies by **TCP**. Know the 3-way handshake (SYN, SYN-ACK, ACK) and why packet loss causes TCP "Retransmissions," which kills API performance.
  * **Diagnostics:**
      * `ip route show`: How does the server know where to send the response?
      * `ss -tunlp`: What is listening on port 443? Is the process actually up?

### Day 3: Web Tech & API Security

**Focus:** The "Bread and Butter" of Apigee.

  * **HTTP Semantics:** Don't just list codes. Group them: `2xx` (Success), `4xx` (Client messed up - think 401/403 for Apigee), `5xx` (Server/Backend messed up - 502/504 are huge for gateways).
  * **The TLS Handshake:** Since Apigee handles secure traffic, you must understand the "Handshake." Focus on: **SNI** (Server Name Indication), Certificate validation, and Cipher negotiation.
  * **Caching & State:** Understand **Cache-Control** headers. Know the "Least Recently Used" (LRU) eviction policy—it's the most common for API gateways.

### Day 4: The "Apigee Integration" Mindset (Mock Prep)

**Focus:** Putting it all together.

  * **Scenario:** A client says, "I'm getting a 502 Bad Gateway from my API."
  * **Your mental flow:**
    1.  **Client/DNS:** Is the domain resolving correctly?
    2.  **Networking:** Can the Gateway reach the Backend? (Ping/Telnet/Curl).
    3.  **Application:** Is the backend process running? (`ss` or `ps`).
    4.  **Resource:** Is the server swapping or out of CPU? (`top`).

## Final "TSE" Tip:

When you don't know an answer, **be systematic.**

Instead of saying "I don't know why the API is slow," say:

> "To diagnose latency, I would start at the edge. I'd check the DNS resolution time, then move to the TCP handshake to see if we have packet loss. If the network is clean, I’d check the Linux server’s IO wait and CPU usage to see if the process is throttled."

That "methodical" approach is often more important to the interviewer than knowing a specific flag for a command. You’ve got this\!
