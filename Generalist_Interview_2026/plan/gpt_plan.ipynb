**TCP vs UDP**
TCP is connection-oriented and provides reliability, ordering, and retransmission. UDP is connectionless and lower overhead but does not guarantee delivery. TCP is better for web traffic where correctness matters. UDP is useful for latency-sensitive use cases like streaming or DNS queries, where some loss may be acceptable.

**Clarify symptom → isolate layer → name commands/tools → likely causes → validate fix**

Example:

* Check connectivity to target host/port
* Check TLS handshake
* Inspect response timing with curl -v
* Look for dependency slowness, firewall, route, or saturation
* Validate after fix with repeated test and latency comparison

# Your answer style in the interview

Use this pattern:

## For theory

**Start simple, then add nuance.**

Example:
“TCP is connection-oriented and reliable, so it is the default fit for web traffic. The tradeoff is overhead from handshake, acknowledgments, and retransmissions. UDP is lighter but gives up delivery guarantees, so it suits use cases like DNS or real-time media where low latency matters more than perfect delivery.”

## For troubleshooting

**Start wide, narrow fast.**

Example:
“If an external API integration starts failing, I’d first determine whether it is a resolution issue, connectivity issue, TLS issue, auth issue, or application-layer response issue. I’d reproduce with curl, inspect DNS resolution, verify the target host and port are reachable, check whether the TLS handshake succeeds, and then inspect the HTTP response code, headers, and payload. From there I’d separate client-side issues like headers or tokens from server-side issues like 5xx errors or rate limiting.”

That is the tone you want.
