@app.route("/cookie-set")
def cookie_set():
    resp = make_response({"message": "cookie set"})
    resp.set_cookie("session_id", "abc123", max_age=300)
    return resp

@app.route("/cookie-check")
def cookie_check():
    session_id = request.cookies.get("session_id")
    if not session_id:
        return {"error": "no session cookie found"}, 400
    return {"message": "cookie received", "session_id": session_id}, 200

@app.route("/headers")
def headers():
    return jsonify(dict(request.headers)), 200

```python
HTTP is stateless by default.
The server does not “remember” the first request unless the client sends something back, such as:

a cookie
a token
a session ID

So:

/cookie-set = server gives you a cookie
/cookie-check = server checks whether client sent it back
```

```python
COOKIES!? For authentication based webapps??

“Because HTTP is stateless, authenticated applications need a mechanism to carry identity across requests. 
Cookies are the most common browser-based method, but bearer tokens in headers are another common approach.”

Other methods exist:
token in Authorization header
local storage / session storage with JS apps
client certificates in rare cases
```
