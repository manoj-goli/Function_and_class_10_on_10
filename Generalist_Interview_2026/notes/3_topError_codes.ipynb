The Success Code
200 OK: This is the standard response for a successful HTTP request
. In an interview, if you're asked how to verify a service is healthy, seeing a "200" is the primary indicator that things are working as expected
.

--------------------------------------------------------------------------------
4xx: Client Errors (The User's Fault)
These codes mean the server thinks the client (browser or app) did something wrong
.
400 Bad Request: The server cannot process the request because the client sent something malformed or incorrect
.
404 Not Found: The most famous code; it means the server cannot find the specific resource (like a webpage or file) the client asked for
.
Note: The following codes (401 and 403) are not explicitly defined in the sources provided, so I am drawing on common technical knowledge. You may want to independently verify these details:
401 Unauthorized: You are not authenticated (the server doesn't know who you are).
403 Forbidden: The server knows who you are, but you don't have permission to access that specific resource.

--------------------------------------------------------------------------------
5xx: Server Errors (The Website's Fault)
These codes mean the server failed to fulfill a valid request
.
500 Internal Server Error: A generic "catch-all" error when the server hits an unexpected condition
. In a high-load situation, a sudden spike in 500-level errors is a major red flag that requires immediate triage
.
Note: The following codes (502 and 503) are also not explicitly detailed in the sources provided, so please verify this information independently:
502 Bad Gateway: Usually happens when one server (acting as a gateway) receives an invalid response from another "upstream" server.
503 Service Unavailable: The server is currently unable to handle the request, often because it is overloaded or down for maintenance.
