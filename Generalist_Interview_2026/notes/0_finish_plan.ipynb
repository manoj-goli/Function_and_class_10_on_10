# Exact interview-ready explanations to memorize

## 1) 502 Bad Gateway

“A 502 usually means the current component is up, but it could not get a valid response from an upstream dependency. I would check whether the upstream is running, listening on the expected port, and reachable.”

## 2) Connection refused

“Connection refused usually means the target host and port were reached, but nothing is listening there, or the service is not accepting connections.”

## 3) 401 vs 403

“401 means authentication is missing or invalid. 403 means the caller is known but not allowed.”

## 4) Self-signed cert

“A self-signed certificate is signed by itself rather than a trusted CA, so clients do not trust it by default.”

## 5) Cookie

“A cookie is state the server asks the client to store and send back on later requests.”

## 6) Cache-Control

“Cache-Control tells clients or intermediaries how a response may be cached, but caching only happens if some client or proxy actually honors it.”
