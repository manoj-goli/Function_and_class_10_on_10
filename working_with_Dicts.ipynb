```python
def process_numbers(nums):
    result = 0
    for i in range(len(nums)):
        if nums[i] % 2:
            result = result + nums[i]
    return result
```

```python
nums = [1, 2, 3, 4, 5]
print(bool(nums[0] % 2))
#print(nums[0])
#basically if there is a remainder when divided by 2 (odd) then % returns true if not (even) then returns false
process_numbers(nums)
```

```python
def count_long_words(words):
    count = 0
    for word in words:
        if len(word) > 3:
            count += 1
    return count
```

```python
words = ["apple", "is", "good", "for", "health", "kantri"]
count_long_words(words)
```

```python
def remove_duplicates(items):
    result = []
    for i in range(len(items)):
        if items[i] not in result:
            result.append(items[i])
    return result


remove_duplicates([2,3,3,4,5,4])
```

```python
def find_max(nums):
    max_num = nums[0]
    for n in nums:
        if n >= max_num:
            max_num = n
    return max_num

find_max([43,-10, -3, -25, 7, -100, 83])
```

```python
def quorb_agg(events):
    out = {}
    for e in events:
        item = e.get("item")
        if item not in out:
            out[item] = out.get(item, [])
        out[item].append(e.get("qty", 1))
    return out
    
    
   # {"item":apple,"qty":2}
    
print(quorb_agg([
    {"item": "apple", "qty": 2},
    {"item": "banana"},
    {"item": "apple", "qty": 3},
    {"qty": 10}
]))
```
