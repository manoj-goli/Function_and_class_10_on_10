```python
def should_alert(lines, window=6, threshold=3):
    recent = deque(maxlen=window)
    for line in lines:
        line = line.strip()
        if not line or line.startswith("#"):
            continue
        recent.append(line)

    errors = 0
    for line in recent:
        parts = line.split()
        if len(parts) < 5:
            continue
        status = int(parts[3])
        if status >= 500:
            errors += 1

    return errors >= threshold, errors


def main(argv):
    if len(argv) != 3:
        raise ValueError("usage: tailer.py <logfile> <window> <threshold>")

    path = argv[0]
    window = int(argv[1])
    threshold = int(argv[2])

    with open(path, "r") as f:
        alert, errors = should_alert(f, window=window, threshold=threshold)

    if alert:
        print(f"ALERT: 5xx={errors} threshold={threshold} window={window}")
    else:
        print(f"OK: 5xx={errors} threshold={threshold} window={window}")
```
