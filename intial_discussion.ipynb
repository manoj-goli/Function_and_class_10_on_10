```python

# Your previous Plain Text content is preserved below:

# # Problem:
# # Given an N x N matrix, rotate it 90 degrees clockwise.
# #
# # Example:
# # Input:
# # [
# #   [1, 2, 3],
# #   [4, 5, 6],
# #   [7, 8, 9]
# # ]
# #
# # Output:
# # [
# #   [7, 4, 1],
# #   [8, 5, 2],
# #   [9, 6, 3]
# # ]
# #
# # You can return a new matrix.
# # No zip(), list comprehension, or numpy - basic Python only

# #inputs - clarification ->Data type-> always a list of lists
# #output - clarification ->Data type-> always a list of lists

# #this patterns is matrix manipulation (2D array)
# #approach: ##clockwise
# #let n = len(matrix)
# #create empty nxn result matrix
# #for every cell(r,c)-->place it all-> result[c][n-1-r]= matrix[r][c]
# #return result

def rotate(matrix):
  n = len(matrix)

  #create an empty nxn matric

  result = []
  for _ in range(n):
    result.append([0] * n)

  #rotate
  for r in range(n):
    for c in range(n):
      result[c][n-1-r]= matrix[r][c]

  return result


print(rotate([
]))

# # Output:
# # [
# #   [7, 4, 1],
# #   [8, 5, 2],
# #   [9, 6, 3]
# # ]
```
