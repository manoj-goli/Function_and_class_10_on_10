```python
class get_unique_count:
    def __init__(self):
        self.uniques = []
    
    def add_view(self,user_id,product_id):
        if (user_id,product_id) not in self.uniques: #(u1,p1) (u1,p2) (u2,p1) (u2,p2)
            self.uniques.append((user_id,product_id)) #(u1,p1) (u1,p2) (u2,p1) (u2,p2)

    def get_unique_count(self, user_id):
        count = 0
        for items in self.uniques: #(u1,p1) (u1,p2) (u2,p1) (u2,p2)
            if user_id == items[0]: #(u1,p1) (u1,p2)
                count += 1
        return count
        

unique_visits = get_unique_count()
unique_visits.add_view("u1", "p1")
unique_visits.add_view("u1", "p1")
unique_visits.add_view("u2", "p1")
print(unique_visits.get_unique_count("u1"))
    
#Big O Notation:
# The time complexity of the add_view method is O(n) because it checks if the (user_id, product_id) tuple is already in the uniques list, which takes O(n) time in the worst case.
# The time complexity of the get_unique_count method is also O(n) because it iterates through the uniques list to count the number of unique products viewed by the specified user_id, which takes O(n) time in the worst case.
# The overall time complexity of the get_unique_count class is O(n) for both methods, where n is the number of unique (user_id, product_id) pairs stored in the uniques list.
# The space complexity of the add_view method is O(n) because it can store up to n unique (user_id, product_id) pairs in the uniques list.
# The space complexity of the get_unique_count class is O(n) because it stores unique (user_id, product_id) pairs in the uniques list, which can grow up to n in size.
# In summary, the get_unique_count class has a time complexity of O(n) for both methods and a space complexity of O(n).

```

```python
class GetUniqueCount:
    def __init__(self):
        # Dictionary where key is user_id, value is a set of product_ids
        self.uniques = {} 
    
    def add_view(self, user_id, product_id):
        if user_id not in self.uniques:
            self.uniques[user_id] = set()
        self.uniques[user_id].add(product_id)

    def get_unique_count(self, user_id):
        # Simply get the length of the set for that specific user
        return len(self.uniques.get(user_id, set()))

unique_visits = GetUniqueCount()
unique_visits.add_view("u1", "p1")
unique_visits.add_view("u1", "p2")
unique_visits.add_view("u2", "p1")

print(unique_visits.get_unique_count("u1")) # Output: 2

#Big O Notation:
# The time complexity of the add_view method is O(1) because adding an element to a set and checking for the existence of a key in a dictionary both take constant time on average.
# The time complexity of the get_unique_count method is O(1) because it retrieves the set of product_ids for the specified user_id and calculates its length, both of which take constant time on average.
# The overall time complexity of the GetUniqueCount class is O(1) for both methods, where n is the number of unique (user_id, product_id) pairs stored in the uniques dictionary.
# The space complexity of the add_view method is O(n) because it can store up to n unique (user_id, product_id) pairs in the uniques dictionary.
# The space complexity of the GetUniqueCount class is O(n) because it stores unique (user_id, product_id) pairs in the uniques dictionary, which can grow up to n in size.
# In summary, the GetUniqueCount class has a time complexity of O(1) for both methods and a space complexity of O(n).
```
