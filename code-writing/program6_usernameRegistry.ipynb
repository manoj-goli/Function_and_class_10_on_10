```python
class UsernameRegistry:
    def __init__(self):
        self.records = set()

    def normalize(self, username: str) -> str:
        return username.strip().casefold()
    
    def register(self, username: str) -> bool:
        if not self.normalize(username):
            return False
        else:
            normalized_username = self.normalize(username)
            if normalized_username not in self.records:
                self.records.add(normalized_username)
                return True
            else:            
                return False
    def remove(self, username: str) -> bool:
        normalized_username = self.normalize(username)
        if normalized_username in self.records:
            self.records.discard(normalized_username)
            return True
        else:
            return False
    
    def is_taken(self, username: str) -> bool:
        return self.normalize(username) in self.records
    def count(self) -> int:
        return len(self.records)
    

```
