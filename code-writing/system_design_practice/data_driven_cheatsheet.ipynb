## Data-Driven Value Differentiation Cheat Sheet

1. **“First, I want to identify the actual goal: is it cost, latency, reliability, or user experience?”**
2. **“The key metric that should drive this decision is ___.”**
3. **“I’d establish a baseline first: before vs after, last deploy, last 30 days, or per region/service.”**
4. **“I would avoid averages if user experience is involved and look at p95 or p99.”**
5. **“I’d segment the data by the biggest dimension that matters: endpoint, tenant, object age, access frequency, or cache hit/miss path.”**
6. **“I want to identify the primary driver, not just the loudest symptom.”**
7. **“If performance already meets SLA, I would prioritize the bigger cost lever.”**
8. **“If most requests are on the miss path, scaling the app tier won’t help much.”**
9. **“I’d choose the option with the highest measurable value improvement for the least engineering effort.”**
10. **“I would prove the change worked by rechecking the same metric after the fix.”**

## 5 filler lines that sound strong in interviews

* **“Users feel tail latency, not averages.”**
* **“I’d start with p95, not mean.”**
* **“Access frequency should determine storage tier.”**
* **“I want correlation with recent changes and traffic patterns.”**
* **“The right answer depends on the dominant driver in the data.”**

## Super short answer template

Use this when you are under pressure:

**Goal:** We are optimizing for ___
**Metric:** The key metric is ___
**Baseline:** I’d compare ___
**Decision:** I’d choose ___ because ___
**Tradeoff:** The tradeoff is ___
**Proof:** I’d validate success by checking ___

## Example skeleton

> We are optimizing for **cost reduction without hurting user experience**.
> The key metric is **access frequency distribution**.
> The baseline shows **most objects are rarely accessed**, while performance already meets SLA.
> So I’d choose **lifecycle tiering**, because it gives the biggest savings for the least effort.
> The tradeoff is **slower retrieval for cold objects**.
> I’d prove it worked by measuring **monthly storage cost and latency on retrievals for cold objects**.
