```python
class LikeTracker:
    def __init__(self):
        self.LikeTrace = set()
    def like(self, user_id, post_id):
        self.LikeTrace.add((user_id, post_id))
        
    def unlike(self, user_id, post_id):
        self.LikeTrace.discard((user_id, post_id))
        
    def has_liked(self, user_id, post_id) -> bool:
        return (user_id, post_id) in self.LikeTrace
        
    def like_count(self, post_id) -> int:
        count = 0
        for items in self.LikeTrace:
            if post_id == items[1]:
                count += 1
        return count
    

# Example usage:
tracker = LikeTracker()
tracker.like("u1", "p1")
tracker.like("u2", "p1")
tracker.like("u3", "p2")

print(tracker.has_liked("u1", "p1"))  # Output: True
print(tracker.has_liked("u2", "p1"))  # Output: True
print(tracker.has_liked("u3", "p2"))  # Output: True
print(tracker.like_count("p1"))  # Output: 2
tracker.unlike("u1", "p1")
print(tracker.has_liked("u1", "p1"))  # Output: False
print(tracker.like_count("p1"))  # Output: 1

#Big O Notation:
# The time complexity of the like and unlike methods is O(1) because they involve adding or removing an element from a set, which is a constant time operation because sets in Python are implemented as hash tables.
# The time complexity of the has_liked method is also O(1) because it checks for the presence of a tuple in the set, which is a constant time operation because of the hash table implementation.
# The time complexity of the like_count method is O(n) where n is the number of likes in the LikeTrace set, because it iterates through all the likes to count how many times a specific post_id appears.
# The space complexity of the LikeTracker class is O(m) where m is the number of unique (user_id, post_id) pairs in the LikeTrace set, because each like is stored as a tuple in the set.
```
