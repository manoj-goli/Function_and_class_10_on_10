We have a social media startup where users can follow each other. We need a basic feature to analyze these connections.

**The Problem:**
Write a function called `get_recommendations` that takes two inputs:
1.  `user_id` (a string/integer representing the current user).
2.  `graph` (a dictionary where the keys are user IDs, and the values are lists of users they follow).

The function should return a list of users that the current user *does not already follow*, but who are followed by at least one of the users that the current user *does* follow.

In other words: **"Recommend the accounts that my followees follow, but I don't follow yet."**

**Example:**
```python
graph = {
    "A": ["B", "C"],
    "B": ["C", "D"],
    "C": ["E"],
    "D": ["F"],
    "E": [],
    "F": []
}

# If the current user is "A":
# - "A" follows ["B", "C"]
# - "B" follows ["C", "D"]
# - "C" follows ["E"]
# - Combined followees of B and C: ["C", "D", "E"]
# - Remove users A already follows: ["B", "C"] 
# - Remove A themselves? (Optional, but we usually filter out self. Let's filter self out for clarity)
# - Result: ["D", "E"]
```

Take your time to read the prompt. Let me know when you are ready to start coding, or if you have any clarifying questions about the input/output format.

````python
# Get the list of users user_id follows from graph[user_id] (call it followees).

# Loop over each person in followees.

# For each followee, get the list of users they follow from graph[followee].

# Collect all of those into a set (to avoid duplicates).

# Remove from this set:

# user_id themselves (optional, but good).

# Any user already in followees.

# Turn the set into a list and return it.



def get_recommendations(user_id, graph):
    recommendations = []
    followees = graph[user_id] # A-->["B","C"]
    for follower in followees:
        followers_follows = graph[follower] #B-->["C","D"]
        
        for person in followers_follows:
            if person not in followees and person != user_id:
                if person not in recommendations:
                    recommendations.append(person)
        
    return recommendations
````

````python
def get_recommendations(user_id, graph):
    followees = graph[user_id]  # who user follows
    recommendations = []  # start with empty list
    
    for follower in followees:
        # Get who this follower follows
        followers_follows = graph[follower]
        
        # Add them to recommendations
        for person in followers_follows:
            # Only add if not already followed and not the user themselves
            if person not in followees and person != user_id:
                if person not in recommendations:  # avoid duplicates
                    recommendations.append(person)
    
    return recommendations
````
