# 8-week roadmap with milestones

## Week 1 — Build the mental model

### Focus

Understand where Kubernetes fits for AI workloads.

### Topics

* training vs inference
* why GPUs need special handling in Kubernetes
* device plugins
* operators
* basic LLM serving flow
* what to observe in GPU-backed workloads

### Milestone

You can clearly explain:

* why Kubernetes is useful for AI inference workloads
* what a device plugin does
* what an operator does
* why this is more about platform/infrastructure than ML theory

You create:

* a `notes/week1-foundations.md`
* a simple architecture sketch
* a short glossary of terms

A 1-page write-up:

> “How Kubernetes brings value to AI serving workloads”

# What each folder is for

## `README.md`

This is the front door to the project.

Include:

* what this repo is
* what you learned
* architecture image
* project phases
* key comparisons
* how to run the main pieces
* what you would improve next

## `docs/`

Higher-level polished docs.

This is where the “showable” content lives:

* summary docs
* capstone
* architecture
* comparison write-up

## `notes/`

Messy learning notes are okay here.

This proves your journey and helps you revise later.

## `manifests/`

All Kubernetes YAML goes here.

Keep it separated by purpose:

* gpu test
* vLLM
* monitoring
* KServe
* Kubeflow

## `scripts/`

Only simple helper scripts.
Not fancy automation. Just enough to:

* create cluster
* deploy things
* clean up

## `experiments/`

This is important.

Store findings like:

* startup time
* latency differences
* deployment friction
* observability differences
* what broke

This makes the repo feel like real engineering work.
