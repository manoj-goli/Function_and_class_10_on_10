```python
from dataclasses import dataclass
from typing import List

# Here are the data classes used by the scheduler.

#reqs:
# ASK 1: single agent can be assigned to zero or more unique tasks based on the task ID
#BUT single TASK cannot be assigned to 2 different agents

#ASK 2: once the task is assigned-->agent1, total hours should not exceed the person/agent's capacity

@dataclass
class Task:
    id: int
    hours: int # the hours required to complete the task

@dataclass
class Agent:
    name: str
    capacity: int # the hours available

# Here is the scheduler that aims to distribute tasks amongst available agents
# and print the results.

def scheduler(tasks: List[Task], team: List[Agent]): #[task1.hours=1, task2.hours=2, task3.hours=5, task4.hours=1], [agent1.capacity=4; agent2.capacity=3; agent3.capacity=2]

    sorted_tasks = sorted(tasks, key=lambda task:task.hours, reverse=True) #(iterable, sort key: lamda kv: kv, reverse)
    #[task1.hours=5, task2.hours=2, task3.hours=1, task4.hours=1]
    
    
    sorted_team = sorted(team, key=lambda agent:agent.capacity, reverse=True)
    #[agent1.capacity=4; agent2.capacity=3; agent3.capacity=2]
    
    #max task hours > max capacity --> print("there are tasks that no agent capacity can take")
    if sorted_tasks[0].hours - sorted_team[0].hours:
      print("there are tasks that no agent capacity can take") # Question 2
      
      # if total sorted_task hours > capacity # Question 3
      
    a = sum(task.hours for task in tasks)  ##similar to max function
    b = sum(agent.capacity for agent in agent)
    
    if a > b:
      print("=====")
    
   
    agent_got_tasks = {} #initiating a empty dict
    assigned = 0 #

    for task in sorted_tasks: #task2.hours=2
        for agent in sorted_team: #agent1.capacity=4
            if agent.capacity - task.hours < 0: #  4 -2 #
                continue
            agent_got_tasks.setdefault(agent.name, []).append(task.id)
            agent.capacity -= task.hours #fixed: Question 1
            assigned += 1
            break

    # task 3 with 5 hours is going to say there are tasks in the next line <--
    if assigned < len(tasks): print("There are still task(s) in the queue") # unassigned = len(tasks) - assigned
    print(f"Agents have the following task assignments: {agent_got_tasks}")

    
# task1.hours=2; task2.hours=3; task3.hours=3; task4.hours=2
# agent1.capacity=; agent2.capacity=3; agent3.capacity=2
# tasks = [task1, task2, task3, task4]
# agents = [agent1, agent2, agent3]
# scheduler(tasks, agents)
```
